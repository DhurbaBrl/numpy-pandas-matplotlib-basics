### Basics

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,50,20)
y=x**2
```

```python
plt.plot(x,y,x,x,y,x) #can plot any number of plots in the same graph like this
plt.show()
```

### label and legend

```python
plt.plot(x,y,label='number-freq')  #for label in legend
plt.xlabel('number')       #to name label on x-axis
plt.ylabel('frequency')    #to name label on y-axis
plt.legend(loc=4)          #loc to change the location of legend
plt.show()
```

### Types of plots  (depends on types of data)

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.hist(x,y)
plt.show()
```

```python
plt.fill(x,y)
```

```python
plt.pie(x=[50,60,70,20],explode=(0,0.08,0,0),labels=['nepal','china','india','cuba'],autopct='%1.1f%%',shadow=True)
plt.show()   #removes the description above the chart

#the slices will be ordered and plotted counter-clockwise
#'x' is the value of each slice
#autopct plots the % inside
```

### Subplots

```python
plt.subplot(2,2,1)      #subplot to make multiple graphs on the same page
plt.plot(x,y)
plt.subplot(2,2,2)
plt.scatter(x,y)
plt.subplot(2,2,3)
plt.hist(x,y)
plt.subplot(2,2,4)
plt.bar(x,y)

plt.tight_layout()      #to remove the overlapping of graphs
```
