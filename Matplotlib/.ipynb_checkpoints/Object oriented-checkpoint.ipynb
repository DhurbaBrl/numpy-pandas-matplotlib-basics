```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,50,20)
y=x**2
```

```python
fig=plt.figure(figsize=(5,4)) #create a virtual plt and fisize in inches inside the tuple
axes=fig.add_axes([0.1,0.1,1,1]) # to adjust the axes origin, width and height
axes.plot(x,y)    #to plot the graph

axes.set_xlabel('x-axis')   
axes.set_ylabel('y-axis')

axes.set_title('First Graph')


plt.show()
```

### Multiple plots

```python
fig=plt.figure()
x1=fig.add_axes([0,0,1,1])
x1.plot(x,y)
x2=fig.add_axes([1,1,0.5,0.5])   # the origin of second graph is at 1,1 position of first graph and its size is half
x2.plot(y,x)
plt.show()
```

### Subplots

```python
fig,axes=plt.subplots(2,2,figsize=(8,8))
axes[(0,0)].plot(x,y)    #tuple inside axes is the dimension in which to plot

axes[(0,0)].set_xlabel('x-axis')   
axes[(0,0)].set_ylabel('y-axis')

axes[(0,0)].set_title('First Graph')



plt.tight_layout()
```

### Styling plots

```python
fig=plt.figure(figsize=(5,4)) 
axes=fig.add_axes([0.1,0.1,1,1])


axes.plot(x,y,color='red',linestyle='-',linewidth=3,alpha=0.5,marker='o',markersize=10)    #alpha is the color intensity 


plt.show()
```

### Set limit

```python
fig=plt.figure(figsize=(5,4)) 
axes=fig.add_axes([0.1,0.1,1,1])

axes.set_xlim(0,40)
                                #to set xlimit and ylimit
axes.set_ylim(0,1500)     

axes.plot(x,y,color='g') 


plt.show()
```
